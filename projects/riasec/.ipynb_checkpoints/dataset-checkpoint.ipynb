```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv("./data.csv", sep="\t")
df_riasec = pd.read_csv("./riasec.csv", sep="\t", header=None)
```

```python
df.head()
```

```python
df_riasec.columns = ["item","text"]
df_riasec.head()
```

```python
# drop unsed cols
df.drop([
    'Unnamed: 93',
     'urban',
     'hand',
     'religion',
     'orientation',
     'race',
     'voted',
     'married',
     'familysize',
     'uniqueNetworkLocation',
     'country',
     'source',
     'major'], axis =1, inplace=True)
```

```python
# remove fast and slow timers (include above 10% and below 95% percentile)
df = df.loc[df.testelapse.between(*df.testelapse.quantile( q = [.05, .95])), :]
```

```python
# remove cases who fail the control items
df = df.loc[~df[["VCL6","VCL9","VCL12"]].any(axis=1), :]
```

```python
# remove cases with any TIPI item missing
df = df.loc[(df.loc[:, [ f"TIPI{col}" for col in range(1,11) ]] > 0).all(axis=1)]
```

```python
# remove cases with low number of high RIASEC items
df = df[(df.loc[:,"R1":"C8"] > 3).sum(axis=1) > 5 ]
```

```python
# remove cases with high number of high RIASEC items
df = df[(df.loc[:,"R1":"C8"] > 3).sum(axis=1) < 37 ]
```

```python
# focus on ages between 18 and 55
df = df.loc[df.age.between(18, 55), :]
```

```python
# focus on participants who declared their gender
df = df.loc[df.gender.between(1,2), :]
```

```python
# subsample female component to match male component
df = pd.concat([
    df.loc[df.gender == 1, :],
    df.loc[df.gender == 2, :].sample(n = df.loc[df.gender == 1, :].shape[0], random_state = 42)
])
```

```python
# reverse TIPI items
tipi_revesed_cols = [ f"TIPI{item}" for item in [2,4,6,8,10] ]
df.loc[:, tipi_revesed_cols ] = 8 - df.loc[:, tipi_revesed_cols ]
```

```python
# score TIPI
scales = dict(extra=[1,6], agrea=[2,7], consc=[3,8], emstab=[4,9], openn=[5,10])

for scale, scale_items in scales.items():
    df[scale] = df[[ f"TIPI{col}" for col in scale_items ]].mean(axis=1)
```

```python
# discretize TIPI scales
for scale in ["extra","agrea","consc","emstab","openn"]:
     df[f"{scale}_q"] = pd.cut(df[scale], bins = [ 0, 4, 7 ], labels=[1,2])
```

```python
# Discretize RIASEC scales
for scale in list("RIASEC"):
    df[f"{scale}"] = df.filter(regex=f"^{scale}\d+$").mean(axis=1)
for scale in list("RIASEC"):
    df[f"{scale}_q"] = pd.cut(df[scale], bins = [ 0, 1, 3, 4, 6, 7  ], labels = range(1,6))
```

```python
# drop unused columns
df.drop(df.filter(regex="^TIPI\d+$", axis=1).columns, inplace=True, axis=1)
df.drop(df.filter(regex="^VCL\d+$", axis=1).columns, inplace=True, axis=1)
df.drop(df.filter(regex="elapse", axis=1).columns, inplace=True, axis=1)
df.drop(["engnat"], axis=1, inplace=True)
```

```python
def fn(x):
    return "\n".join(df_riasec.loc[(x == 5).values, "text"])

df["text"] = df.loc[:, "R1":"C8"].apply(lambda x: fn(x) , axis = 1)
```

```python
df_final = df[["gender","text","agrea_q"]]
df_final = df_final[df_final["text"].str.len() > 0]
```
