```python
import numpy as np
import pandas as pd

pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
items = pd.read_csv("items.csv", header=None)
items.columns = ["item","text"]
items.head()
```

```python
df = pd.read_csv("data.csv", sep="\t")
df.head()
```

```python
# reorder answers
def fn(x):
    values, idx = x.values[::2], x.values[1::2]
    return pd.Series(values, index=idx).sort_index()

# add reordered answers to dataframe
df = pd.concat([
        df,
        pd.DataFrame(
            data = df.filter(regex="^Q\d*[A|I]$").apply(fn, axis=1).values, 
            columns=[ f"Q{i}AR" for i in range(1,92) ]
        )
    ], axis = 1
)

df
```

```python
# drop unused columns
df = df.drop(df.filter(regex="^Q\d*[I|E]$").columns, axis = 1)
df = df.drop(["dateload","country","engnat"], axis = 1)
```

```python
# drop slow timers and fast timers
df = df[df.introelapse.between(*df.introelapse.quantile(q=[.05, .95]))]
df = df[df.testelapse.between(*df.testelapse.quantile(q=[.05, .95]))]
df = df[df.surveyelapse.between(*df.surveyelapse.quantile(q=[.05, .95]))]
df.shape
```

```python
# drop cases with self report of IE missing
df = df[df.IE.between(1,3)]
df.shape
```

```python
# drop cases without high scores (>3)
df = df.loc[(df.filter(regex="^Q\d*AR$") > 3).sum(axis=1) > 9, :]
df.shape
```

```python
# for each case, build text by joining items whose likert answer is == 4 or == 5 
df["text"] = df.filter(regex="^Q\d*AR$")\
    .apply(lambda x: "\n".join(items.loc[ x.values > 3 , "text"]) , axis=1)
```
